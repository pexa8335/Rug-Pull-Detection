```python
import ast
import re
# Danh sách 4 file txt
files = [
    '/kaggle/input/rawdatajson/trainai.txt',
    '/kaggle/input/rawdatajson/trainai2.txt',
    '/kaggle/input/rawdatajson/trainai3.txt',
    '/kaggle/input/rawdatajson/trainai4.txt'
]
```

```python
all_data = []

for file in files:
    with open(file, 'r') as f:
        for line in f:
            line = line.strip()
            if line:
                # Tìm phần dict từ "Data: {....}"
                match = re.search(r'Data:\s*(\{.*\})$', line)
                if match:
                    dict_str = match.group(1)
                    
                    # Sửa lỗi TRUE/FALSE (chuyển về Python syntax)
                    dict_str = dict_str.replace('TRUE', 'True').replace('FALSE', 'False')
                    
                    try:
                        data_dict = ast.literal_eval(dict_str)
                        all_data.append(data_dict)
                    except Exception as e:
                        print(f"Lỗi parse dict: {dict_str[:100]}... -> {e}")
                else:
                    print(f"Không tìm thấy dict trong dòng: {line[:80]}...")

# Tạo DataFrame
df = pd.DataFrame(all_data)

# Kiểm tra kết quả
print(f"Số dòng hợp lệ: {len(df)}")
df.head()
```

```python
X_train = pd.read_csv(path)
X_train
```

```python
import joblib
best_model = joblib.load ('/kaggle/input/isofo/other/default/1/isolation_forest_model.joblib')
```

```python
col_to_drop = [
    'FIRST_POOL_ACTIVITY_TIMESTAMP_ts',
    'LAST_POOL_ACTIVITY_TIMESTAMP_ts',
    'LAST_SWAP_TIMESTAMP_ts',
    'LIQUIDITY_POOL_ADDRESS',
    'MINT'
]
X_train = X_train.drop(columns = col_to_drop)
```

```python
X_train.columns
```

```python
# Gọi decision function trên toàn bộ dữ liệu đã dùng fit
scores = best_model.decision_function(X_train)

# Tạo một DataFrame chứa X và score
df_scores = X_train.copy()
df_scores['anomaly_score'] = scores  # cao là normal, thấp là anomaly
```

```python
threshold = np.percentile(scores, 95)  
normal_candidates_df = df_scores[df_scores['anomaly_score'] >= threshold]
normal_candidates_df
```

```python
df.sample(5)
```

```python
normal_candidates_df = normal_candidates_df.copy()
normal_candidates_df['true_label'] = False
normal_candidates_df = normal_candidates_df.drop(columns = 'anomaly_score')
normal_candidates_df
```

```python
columns_to_drop = ['anomaly_score_isolation_forest', 
                   'is_outlier_isolation_forest_label', 
                   'anomaly_score',     'FIRST_POOL_ACTIVITY_TIMESTAMP_ts',
    'LAST_POOL_ACTIVITY_TIMESTAMP_ts',
    'LAST_SWAP_TIMESTAMP_ts',
    'LIQUIDITY_POOL_ADDRESS',
    'MINT'
]

df = df.drop(columns=columns_to_drop, errors='ignore')
```

```python
print("df shape:", df.shape)
print("normal_candidates_df shape:", normal_candidates_df.shape)
df
```

```python
label_df = pd.concat([df, normal_candidates_df], ignore_index=True)
label_df.head()
```

```python
label_df
```

```python
features = label_df.drop(columns=[
    'true_label', 
    'anomaly_score_isolation_forest', 
    'is_outlier_isolation_forest_label', 
    'anomaly_score'], 
    errors='ignore')
```

```python
label_df['is_outlier_isolation_forest'] = best_model.predict(features)
label_df
```

```python
y_true = label_df['true_label'].astype(bool)
y_pred = label_df['is_outlier_isolation_forest'] == -1
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

print("Accuracy:", accuracy_score(y_true, y_pred))
print("Precision:", precision_score(y_true, y_pred))
print("Recall:", recall_score(y_true, y_pred))
print("F1-score:", f1_score(y_true, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_true, y_pred))

print("\nClassification Report:")
print(classification_report(y_true, y_pred))
```
