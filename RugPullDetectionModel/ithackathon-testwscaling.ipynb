```python
all_data = []

for file in files:
    with open(file, 'r') as f:
        for line in f:
            line = line.strip()
            if line:
                match = re.search(r'Data:\s*(\{.*\})$', line)
                if match:
                    dict_str = match.group(1)
                    dict_str = dict_str.replace('TRUE', 'True').replace('FALSE', 'False')
                    try:
                        data_dict = ast.literal_eval(dict_str)
                        all_data.append(data_dict)
                    except Exception as e:
                        print(f"Error parsing dict: {dict_str[:100]}... -> {e}")
                else:
                    print(f"Can find dict in line: {line[:80]}...")

df_labelled = pd.DataFrame(all_data)

# Kiá»m tra káº¿t quáº£
print(f"Valid rows: {len(df_labelled)}")
df_labelled
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, roc_curve, auc, fbeta_score, make_scorer

# X_test_result['true_label'] Äang lÃ  -1 (anomaly) vÃ  1 (normal)
# X_test_result['anomaly_score'] lÃ  Äiá»m tá»« decision_function (cÃ ng tháº¥p cÃ ng báº¥t thÆ°á»ng)

true_labels_original = X_test_result['true_label'].copy() # -1 for anomaly, 1 for normal
anomaly_scores = X_test_result['anomaly_score'].copy()

# convert true_label 0/1
# vÃ  Äiá»u chá»nh scores sao cho Äiá»m CAO HÆ N tÆ°Æ¡ng á»©ng vá»i lá»p POSITIVE (anomaly)
# Positive is class anomaly (-1)
y_true_binary = (true_labels_original == -1).astype(int) # 1 for anomaly, 0 for normal

# Scores: IsolationForest cho Äiá»m tháº¥p hÆ¡n cho anomaly.
# roc_curve vÃ  precision_recall_curve mong Äá»£i Äiá»m cao hÆ¡n cho lá»p positive.
# VÃ¬ váº­y, chÃºng ta Äáº£o dáº¥u anomaly_scores.
y_scores_inverted = -anomaly_scores # BÃ¢y giá» Äiá»m cao hÆ¡n cÃ³ nghÄ©a lÃ  kháº£ nÄng anomaly cao hÆ¡n
```

```python
precision, recall, thresholds_pr = precision_recall_curve(y_true_binary, y_scores_inverted, pos_label=1)

# thresholds_pr lÃ  cÃ¡c ngÆ°á»¡ng trÃªn y_scores_inverted.
# Äá» Ã¡p dá»¥ng cho anomaly_scores gá»c, chÃºng ta cáº§n Äáº£o dáº¥u láº¡i: -thresholds_pr

# Loáº¡i bá» giÃ¡ trá» cuá»i cÃ¹ng cá»§a precision vÃ  recall Äá» khá»p vá»i kÃ­ch thÆ°á»c cá»§a thresholds_pr
# (precision_recall_curve tráº£ vá» precision/recall cÃ³ kÃ­ch thÆ°á»c N+1, thresholds cÃ³ N)
if len(precision) > len(thresholds_pr):
    precision = precision[:-1]
    recall = recall[:-1]

plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label='Precision-Recall Curve')
plt.xlabel('Recall (for Anomaly class -1)')
plt.ylabel('Precision (for Anomaly class -1)')
plt.title('Precision-Recall Curve for Anomaly Detection')
plt.legend()
plt.grid(True)
plt.show()

# TÃ¬m ngÆ°á»¡ng Äá» tá»i Æ°u recall (vÃ­ dá»¥: tÃ¬m recall gáº§n 0.7, 0.8, 0.9...)
# Hoáº·c tÃ¬m Äiá»m gáº§n gÃ³c trÃªn bÃªn pháº£i cá»§a Äá» thá» PR (náº¿u muá»n cÃ¢n báº±ng hÆ¡n)
# VÃ­ dá»¥: tÃ¬m ngÆ°á»¡ng cho recall >= 0.8
desired_recall = 0.8
candidate_indices = np.where(recall >= desired_recall)[0]
if len(candidate_indices) > 0:
    # Trong sá» cÃ¡c ngÆ°á»¡ng Äáº¡t recall mong muá»n, chá»n ngÆ°á»¡ng cÃ³ precision cao nháº¥t
    best_idx_for_desired_recall = candidate_indices[np.argmax(precision[candidate_indices])]
    
    optimal_threshold_pr_inverted = thresholds_pr[best_idx_for_desired_recall]
    optimal_threshold_for_scores = -optimal_threshold_pr_inverted # Äáº£o dáº¥u láº¡i cho anomaly_score gá»c
    
    print(f"NgÆ°á»¡ng tá»« PR curve cho recall >= {desired_recall}: {optimal_threshold_for_scores:.4f}")
    print(f"  Vá»i Recall: {recall[best_idx_for_desired_recall]:.4f}, Precision: {precision[best_idx_for_desired_recall]:.4f}")
else:
    print(f"KhÃ´ng tÃ¬m tháº¥y ngÆ°á»¡ng nÃ o cho recall >= {desired_recall}")
    # Náº¿u khÃ´ng Äáº¡t ÄÆ°á»£c recall mong muá»n, cÃ³ thá» láº¥y ngÆ°á»¡ng cho recall cao nháº¥t
    if len(recall) > 0 :
        max_recall_idx = np.argmax(recall)
        optimal_threshold_pr_inverted = thresholds_pr[max_recall_idx]
        optimal_threshold_for_scores = -optimal_threshold_pr_inverted
        print(f"NgÆ°á»¡ng cho recall cao nháº¥t cÃ³ thá»: {optimal_threshold_for_scores:.4f}")
        print(f"  Vá»i Recall: {recall[max_recall_idx]:.4f}, Precision: {precision[max_recall_idx]:.4f}")
    else:
        optimal_threshold_for_scores = np.min(anomaly_scores) - 1 # Fallback: classify everything as anomaly
        print(f"KhÃ´ng cÃ³ Äiá»m nÃ o trÃªn PR curve, fallback threshold: {optimal_threshold_for_scores:.4f}")
```
