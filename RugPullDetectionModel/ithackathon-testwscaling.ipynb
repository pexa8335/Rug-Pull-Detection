```python
import ast
import re
# 4 file has label for anomaly 
files = [
    '/kaggle/input/rawdatajson/trainai.txt',
    '/kaggle/input/rawdatajson/trainai2.txt',
    '/kaggle/input/rawdatajson/trainai3.txt',
    '/kaggle/input/rawdatajson/trainai4.txt'
]
```

```python
all_data = []

for file in files:
    with open(file, 'r') as f:
        for line in f:
            line = line.strip()
            if line:
                match = re.search(r'Data:\s*(\{.*\})$', line)
                if match:
                    dict_str = match.group(1)
                    dict_str = dict_str.replace('TRUE', 'True').replace('FALSE', 'False')
                    try:
                        data_dict = ast.literal_eval(dict_str)
                        all_data.append(data_dict)
                    except Exception as e:
                        print(f"Error parsing dict: {dict_str[:100]}... -> {e}")
                else:
                    print(f"Can find dict in line: {line[:80]}...")

df_labelled = pd.DataFrame(all_data)

# Kiểm tra kết quả
print(f"Valid rows: {len(df_labelled)}")
df_labelled
```

```python
X_train = pd.read_csv(path)
X_train
```

```python
import joblib
best_model = joblib.load ('/kaggle/input/isofo/other/default/1/isolation_forest_model.joblib')
```

```python
col_to_drop = [
    'FIRST_POOL_ACTIVITY_TIMESTAMP_ts',
    'LAST_POOL_ACTIVITY_TIMESTAMP_ts',
    'LAST_SWAP_TIMESTAMP_ts',
    'LIQUIDITY_POOL_ADDRESS',
    'MINT'
    ]
X_train = X_train.drop(columns = col_to_drop)
```

```python
X_train
```

```python
#  Transform data = fit scaler
X_scaled = scaler.transform(X_train)

#  Convert to DataFrame 
X_scaled = pd.DataFrame(X_scaled, columns=X_train.columns)
```

```python
 X_scaled
```

```python
# Gọi decision function trên toàn bộ dữ liệu đã dùng fit
scores = best_model.decision_function(X_scaled)

threshold = np.percentile(scores, 95)

y_normal = (scores >= threshold).astype(int)
```

```python
# Lọc trực tiếp ra những dòng có score >= threshold
normal_df_for_test = X_train[scores >= threshold].copy()

# (Tuỳ chọn) Gán cột anomaly_score để kiểm tra
normal_df_for_test['anomaly_score'] = scores[scores >= threshold]
```

```python
true_label = X_test['true_label'].copy()  # lưu lại nhãn

# Bỏ cột này trước khi scale
X_test_features = X_test.drop(columns=['true_label'])

# Scale feature columns
X_test_scaled = scaler.transform(X_test_features)
```
