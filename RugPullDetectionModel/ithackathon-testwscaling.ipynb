```python
all_data = []

for file in files:
    with open(file, 'r') as f:
        for line in f:
            line = line.strip()
            if line:
                match = re.search(r'Data:\s*(\{.*\})$', line)
                if match:
                    dict_str = match.group(1)
                    dict_str = dict_str.replace('TRUE', 'True').replace('FALSE', 'False')
                    try:
                        data_dict = ast.literal_eval(dict_str)
                        all_data.append(data_dict)
                    except Exception as e:
                        print(f"Error parsing dict: {dict_str[:100]}... -> {e}")
                else:
                    print(f"Can find dict in line: {line[:80]}...")

df_labelled = pd.DataFrame(all_data)

# Kiểm tra kết quả
print(f"Valid rows: {len(df_labelled)}")
df_labelled
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, roc_curve, auc, fbeta_score, make_scorer

# X_test_result['true_label'] đang là -1 (anomaly) và 1 (normal)
# X_test_result['anomaly_score'] là điểm từ decision_function (càng thấp càng bất thường)

true_labels_original = X_test_result['true_label'].copy() # -1 for anomaly, 1 for normal
anomaly_scores = X_test_result['anomaly_score'].copy()

# convert true_label 0/1
# và điều chỉnh scores sao cho điểm CAO HƠN tương ứng với lớp POSITIVE (anomaly)
# Positive is class anomaly (-1)
y_true_binary = (true_labels_original == -1).astype(int) # 1 for anomaly, 0 for normal

# Scores: IsolationForest cho điểm thấp hơn cho anomaly.
# roc_curve và precision_recall_curve mong đợi điểm cao hơn cho lớp positive.
# Vì vậy, chúng ta đảo dấu anomaly_scores.
y_scores_inverted = -anomaly_scores # Bây giờ điểm cao hơn có nghĩa là khả năng anomaly cao hơn
```

```python
precision, recall, thresholds_pr = precision_recall_curve(y_true_binary, y_scores_inverted, pos_label=1)

# thresholds_pr là các ngưỡng trên y_scores_inverted.
# Để áp dụng cho anomaly_scores gốc, chúng ta cần đảo dấu lại: -thresholds_pr

# Loại bỏ giá trị cuối cùng của precision và recall để khớp với kích thước của thresholds_pr
# (precision_recall_curve trả về precision/recall có kích thước N+1, thresholds có N)
if len(precision) > len(thresholds_pr):
    precision = precision[:-1]
    recall = recall[:-1]

plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label='Precision-Recall Curve')
plt.xlabel('Recall (for Anomaly class -1)')
plt.ylabel('Precision (for Anomaly class -1)')
plt.title('Precision-Recall Curve for Anomaly Detection')
plt.legend()
plt.grid(True)
plt.show()

# Tìm ngưỡng để tối ưu recall (ví dụ: tìm recall gần 0.7, 0.8, 0.9...)
# Hoặc tìm điểm gần góc trên bên phải của đồ thị PR (nếu muốn cân bằng hơn)
# Ví dụ: tìm ngưỡng cho recall >= 0.8
desired_recall = 0.8
candidate_indices = np.where(recall >= desired_recall)[0]
if len(candidate_indices) > 0:
    # Trong số các ngưỡng đạt recall mong muốn, chọn ngưỡng có precision cao nhất
    best_idx_for_desired_recall = candidate_indices[np.argmax(precision[candidate_indices])]
    
    optimal_threshold_pr_inverted = thresholds_pr[best_idx_for_desired_recall]
    optimal_threshold_for_scores = -optimal_threshold_pr_inverted # Đảo dấu lại cho anomaly_score gốc
    
    print(f"Ngưỡng từ PR curve cho recall >= {desired_recall}: {optimal_threshold_for_scores:.4f}")
    print(f"  Với Recall: {recall[best_idx_for_desired_recall]:.4f}, Precision: {precision[best_idx_for_desired_recall]:.4f}")
else:
    print(f"Không tìm thấy ngưỡng nào cho recall >= {desired_recall}")
    # Nếu không đạt được recall mong muốn, có thể lấy ngưỡng cho recall cao nhất
    if len(recall) > 0 :
        max_recall_idx = np.argmax(recall)
        optimal_threshold_pr_inverted = thresholds_pr[max_recall_idx]
        optimal_threshold_for_scores = -optimal_threshold_pr_inverted
        print(f"Ngưỡng cho recall cao nhất có thể: {optimal_threshold_for_scores:.4f}")
        print(f"  Với Recall: {recall[max_recall_idx]:.4f}, Precision: {precision[max_recall_idx]:.4f}")
    else:
        optimal_threshold_for_scores = np.min(anomaly_scores) - 1 # Fallback: classify everything as anomaly
        print(f"Không có điểm nào trên PR curve, fallback threshold: {optimal_threshold_for_scores:.4f}")
```
