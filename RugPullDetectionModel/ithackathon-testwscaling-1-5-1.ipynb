```python
import ast
import re
# 4 file has label for anomaly 
files = [
    '/kaggle/input/rawdatajson/trainai.txt',
    '/kaggle/input/rawdatajson/trainai2.txt',
    '/kaggle/input/rawdatajson/trainai3.txt',
    '/kaggle/input/rawdatajson/trainai4.txt'
]
```

```python
all_data = []

for file in files:
    with open(file, 'r') as f:
        for line in f:
            line = line.strip()
            if line:
                match = re.search(r'Data:\s*(\{.*\})$', line)
                if match:
                    dict_str = match.group(1)
                    dict_str = dict_str.replace('TRUE', 'True').replace('FALSE', 'False')
                    try:
                        data_dict = ast.literal_eval(dict_str)
                        all_data.append(data_dict)
                    except Exception as e:
                        print(f"Error parsing dict: {dict_str[:100]}... -> {e}")
                else:
                    print(f"Can find dict in line: {line[:80]}...")

df_labelled = pd.DataFrame(all_data)

# Kiểm tra kết quả
print(f"Valid rows: {len(df_labelled)}")
df_labelled
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, roc_curve, auc, fbeta_score, make_scorer

# X_test_result['true_label'] currently is -1 (anomaly) & 1 (normal)


true_labels_original = X_test_result['true_label'].copy() # -1 for anomaly, 1 for normal
anomaly_scores = X_test_result['anomaly_score'].copy()

# These function (precision recall curve need binary type)
y_true_binary = (true_labels_original == -1).astype(int) # 1 for anomaly, 0 for normal

# Scores: IsolationForest give lower scores for anomaly
# roc_curve và precision_recall_curve expect high score for positive (anomaly label)
y_scores_inverted = -anomaly_scores 
```

```python
precision, recall, thresholds_pr = precision_recall_curve(y_true_binary, y_scores_inverted, pos_label=1)


# Loại bỏ giá trị cuối cùng của precision và recall để khớp với kích thước của thresholds_pr
# (precision_recall_curve trả về precision/recall có kích thước N+1, thresholds có N)
if len(precision) > len(thresholds_pr):
    precision = precision[:-1]
    recall = recall[:-1]

plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.', label='Precision-Recall Curve')
plt.xlabel('Recall (for Anomaly class -1)')
plt.ylabel('Precision (for Anomaly class -1)')
plt.title('Precision-Recall Curve for Anomaly Detection')
plt.legend()
plt.grid(True)
plt.show()

desired_recall = 0.8
candidate_indices = np.where(recall >= desired_recall)[0]
if len(candidate_indices) > 0:
    best_idx_for_desired_recall = candidate_indices[np.argmax(precision[candidate_indices])]
    
    optimal_threshold_pr_inverted = thresholds_pr[best_idx_for_desired_recall]
    optimal_threshold_for_scores = -optimal_threshold_pr_inverted # Đảo dấu lại cho anomaly_score gốc
    
    print(f"Threshold of PR curve for recall >= {desired_recall}: {optimal_threshold_for_scores:.4f}")
    print(f"  With Recall: {recall[best_idx_for_desired_recall]:.4f}, Precision: {precision[best_idx_for_desired_recall]:.4f}")
else:
    print(f"Can't find recall >= {desired_recall}")
    # Nếu không đạt được recall mong muốn, có thể lấy ngưỡng cho recall cao nhất
    if len(recall) > 0 :
        max_recall_idx = np.argmax(recall)
        optimal_threshold_pr_inverted = thresholds_pr[max_recall_idx]
        optimal_threshold_for_scores = -optimal_threshold_pr_inverted
        print(f"Best recall possible: {optimal_threshold_for_scores:.4f}")
        print(f"  With Recall: {recall[max_recall_idx]:.4f}, Precision: {precision[max_recall_idx]:.4f}")
    else:
        optimal_threshold_for_scores = np.min(anomaly_scores) - 1 # Fallback: classify everything as anomaly
        print(f"Không có điểm nào trên PR curve, fallback threshold: {optimal_threshold_for_scores:.4f}")
```

```python
from sklearn.metrics import classification_report, fbeta_score

# Tính F2-score và classification report theo từng threshold
thresholds_f2 = np.sort(np.unique(anomaly_scores)) 
f2_scores = []
all_crs = []

for th in thresholds_f2:
    predicted_labels_temp = np.where(anomaly_scores < th, -1, 1)
    f2 = fbeta_score(true_labels_original, predicted_labels_temp, beta=2, pos_label=-1, zero_division=0)
    f2_scores.append(f2)

    cr = classification_report(true_labels_original, predicted_labels_temp, output_dict=True, zero_division=0)
    all_crs.append({
        'threshold': th,
        'f2': f2,
        'precision': cr['-1']['precision'] if '-1' in cr else 0,
        'recall': cr['-1']['recall'] if '-1' in cr else 0
    })

# Lọc các threshold có recall < 1.0
valid_crs = [cr for cr in all_crs if cr['recall'] < 1.0]


valid_crs_sorted = sorted(valid_crs, key=lambda x: x['f2'], reverse=True)

if len(valid_crs_sorted) >= 2:
    selected = valid_crs_sorted[50]  


# In thông tin threshold
print(f"Threshold: {selected['threshold']:.4f}")
print(f"  F2-score: {selected['f2']:.4f}")
print(f"  Precision: {selected['precision']:.4f}")
print(f"  Recall: {selected['recall']:.4f}")

# Classification Report
predicted_labels_f2_optimized = np.where(anomaly_scores < selected['threshold'], -1, 1)
print("\nClassification Report:")
print(classification_report(true_labels_original, predicted_labels_f2_optimized))
```
