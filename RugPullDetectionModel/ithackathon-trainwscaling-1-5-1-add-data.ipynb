```python
import optuna
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split

# estimated_contamination = 0.005 # Báº¡n cáº§n Äáº£m báº£o biáº¿n nÃ y ÄÃ£ ÄÆ°á»£c Äá»nh nghÄ©a á» ÄÃ¢u ÄÃ³ trong code cá»§a báº¡n

def objective(trial, X_data):
    n_features_in_data = X_data.shape[1]
    n_estimators = trial.suggest_int('n_estimators', 50, 400)
    max_samples_upper_bound = min(2048, len(X_data))
    max_samples = trial.suggest_int('max_samples', 100, max_samples_upper_bound)
    max_features_ratio = trial.suggest_float('max_features_ratio', 0.1, 1.0)
    bootstrap = trial.suggest_categorical('bootstrap', [True, False])


    contamination_opt = trial.suggest_float('contamination', 0.001, 0.05, log=True) 
    # Tuy nhiÃªn, viá»c tá»i Æ°u 'contamination' cáº§n cáº©n tháº­n vÃ¬ nÃ³ áº£nh hÆ°á»ng trá»±c tiáº¿p Äáº¿n ngÆ°á»¡ng phÃ¢n loáº¡i.
    # Hiá»n táº¡i, chÃºng ta giá»¯ nguyÃªn giÃ¡ trá» estimated_contamination báº¡n ÄÃ£ cung cáº¥p.
    # Náº¿u báº¡n bá» comment dÃ²ng trÃªn, hÃ£y thay tháº¿ `contamination=estimated_contamination` báº±ng `contamination=contamination_opt`

    model = IsolationForest(
        n_estimators=n_estimators,
        max_samples=max_samples,
        contamination=contamination_opt,
        max_features=max_features_ratio,    
        bootstrap=bootstrap,               
        random_state=trial.number,         
        n_jobs=-1                            
    )

    model.fit(X_data)
    anomaly_scores = model.decision_function(X_data)

    inlier_scores = anomaly_scores[anomaly_scores > 0]
    outlier_scores = anomaly_scores[anomaly_scores <= 0]

    median_inlier = np.median(inlier_scores) if len(inlier_scores) > 0 else 0
    median_outlier = np.median(outlier_scores) if len(outlier_scores) > 0 else 0
    objective_value = median_inlier - median_outlier

    return -objective_value
```

```python
columns_to_drop_for_X = [
    'LIQUIDITY_POOL_ADDRESS',
    'MINT', 
    'LAST_SWAP_TX_ID', 
    'LAST_POOL_ACTIVITY_TIMESTAMP_ts',
    'FIRST_POOL_ACTIVITY_TIMESTAMP_ts',
    'LAST_SWAP_TIMESTAMP_ts'
]
X = df.drop(columns=columns_to_drop_for_X, errors='ignore')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.head()
```

```python
study = optuna.create_study(direction='minimize')
study.optimize(lambda trial: objective(trial, X_scaled), n_trials=100) 
# In ra káº¿t quáº£ tá»t nháº¥t
print("\nOptimization result:")
print(f"Best params: {study.best_params}")
print(f"The best objective function value (negative of the median difference): {study.best_value}")
```

```python
best_params_from_optuna = study.best_params
print(f"Best params from optuna: {best_params_from_optuna}")

best_model = IsolationForest(
    n_estimators=best_params_from_optuna['n_estimators'],
    max_samples=best_params_from_optuna['max_samples'],
    contamination=best_params_from_optuna['contamination'], 
    max_features=best_params_from_optuna.get('max_features_ratio', 1.0),
    bootstrap=best_params_from_optuna.get('bootstrap', False),          
    random_state=42,  
    n_jobs=-1         
)

best_model.fit(X_scaled)
best_model.get_params()
```

```python
anomaly_scores_tuned = best_model.decision_function(X_scaled)
predictions_tuned = best_model.predict(X_scaled)
```

```python
df['anomaly_score_isolation_forest'] = anomaly_scores_tuned
df['is_outlier_isolation_forest'] = predictions_tuned
```
